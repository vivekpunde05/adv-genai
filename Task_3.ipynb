```python
#Problem Statement 1: Employee Performance Bonus Eligibility
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
#find  the highest score
highest_score = max(employees.values())

#find employees with highest score
top_performers = []
for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

#display result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
#Problem Statement 2: Search Query Keyword Analysis
import string
query = "Buy mobile phone buy phone online"

#convet to lowercase
query= query.lower()

#remove punctuation
query = query.translate(str.maketrans('','', string.punctuation))

#split into words
words = query.split()

#count frequency
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] +=1
    else:
        word_count[word] = 1

#Filter words with frequency > 1
result = {}

for word, count in word_count.items():
    if count > 1:
        result[word] = count

print(result)
```

```python
#Problem Statement 3: Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:   # Check if even
        valid_readings.append((i, sensor_readings[i]))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#Problem Statement 4: Email Domain Usage Analysis
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_count = {}

#Extract domain and count
for email in emails:
    domain = email.split("@")[1]
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

#Calculate percentage
total_emails = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
#Problem Statement 5: Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]

total_sales = 0
for value in sales:
    total_sales += value

#Calculate average
average_sales = total_sales / len(sales)

#Calculate 30% above average threshold
threshold = average_sales * 1.30

#Detect sales spikes
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
# Problem Statement 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

for uid in user_ids:
    if uid in id_count:
        id_count[uid] += 1  
    else:
        id_count[uid] = 1

for uid, count in id_count.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
