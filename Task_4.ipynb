```python
#Problem 1: Mobile Recharge Validation System
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        try:
            amount = int(input("Enter recharge amount: ₹"))

            if amount < 50:
                print("Recharge amount must be greater than ₹50.")
            elif amount not in valid_plans:
                print("Invalid plan selected. Valid plans are:", valid_plans)
            else:
                print("Recharge Successful.")
                break

        except ValueError:
            print("Please enter a valid number.")
validate_recharge()

```

```python
#Problem 2: Inventory Reorder Alert System
def inventory_alert():
    products = {}
    n = int(input("Enter number of products: "))

    for i in range(n):
        name = input("Enter product name: ")
        stock = int(input("Enter stock quantity: "))
        products[name] = stock

    print("\nInventory Status:")
    for product, stock in products.items():
        if stock < 15:
            print(product, "→ Reorder Alert")
        else:
            print(product, "→ Stock OK")

inventory_alert()
```

```python
#Problem 3: Student Result Processing System
def student_result():
    marks = []
    n = int(input("Enter number of subjects: "))

    for i in range(n):
        mark = float(input(f"Enter marks for subject {i+1}: "))
        marks.append(mark)

    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)
    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

result = student_result()
print("Result:", result)
```

```python
#Problem 4: Cab Fare Estimator with Retry Option
def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12

    fare = base_fare + (distance * per_km)

    if peak:
        fare += fare * 0.25

    return fare


while True:
    try:
        km = float(input("Enter distance in km: "))
        peak_input = input("Is it peak hour? (yes/no): ").lower()

        if peak_input == "yes":
            peak = True
        else:
            peak = False

        total_fare = calculate_fare(km, peak)
        print("Total Fare: ₹", total_fare)
        break

    except ValueError:
        print("Invalid input. Please try again.")
```

```python
#Problem 5: Employee Attendance Eligibility Checker
def attendance_eligibility():
    attendance = []
    n = int(input("Enter total number of days: "))

    for i in range(n):
        status = input(f"Day {i+1} (P/A): ").upper()
        attendance.append(status)

    present_days = 0
    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100
    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

status = attendance_eligibility()
print("Status:", status)
```

```python
#Problem 6: Password Strength Checker
def password_checker():
    password = input("Enter password: ")

    special_chars = "@#$"
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak Password (Minimum length 8 required)"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and special character)"

print(password_checker())
```
