```python
# Problem 1: Social Media – Post Engagement Analyzer
def analyze_post_engagement(likes_list):
    total_likes = 0
    
    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes
    
    print("Total Likes:", total_likes)
    
    # Check post status
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example usage
likes_data = [200, 150, 300, 400]   # Sample likes from different sources
analyze_post_engagement(likes_data)
```

```python
# Problem 2: Healthcare – Medicine Stock Alert System

def check_medicine_stock(stock):
    
    if stock < 10:
        return "Low Stock Alert"
    else:
        return "Stock Sufficient"


# Taking user input
medicine_stock = int(input("Enter Medicine Stock: "))

status = check_medicine_stock(medicine_stock)

print("Medicine Stock:", medicine_stock)
print("Status:", status)
```

```python
#Problem 3: Agriculture – Rainfall Adequacy Checker
def rainfall_checker(rainfall_data, required_level):
    total = sum(rainfall_data)
    avg = total/len(rainfall_data)
    print("Average Rainfall:", avg)
    # Check adequacy
    if avg >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

season_rainfall = list(map(int,input("Enter Seasonal Rainfall data (in mm): ").split()))
required_rainfall = int(input("required_rainfall: "))
rainfall_checker(season_rainfall, required_rainfall)
            
```

```python
# Problem 4: Duplicate Account Detection

def check_duplicate_accounts(usernames):
    
    unique_users = set(usernames)   # Set removes duplicates
    
    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example usernames list
accounts = ["vivek01", "rahul22", "anita09", "vivek01"]

check_duplicate_accounts(accounts)
```

```python
# Problem 7: System Monitoring – Application Health Checker

def check_application_health(error_count):
    
    print("Error Count:", error_count)
    
    if error_count == 0:
        print("System Status: Healthy")
    elif error_count <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# Taking user input
errors = int(input("Enter Error Count: "))

check_application_health(errors)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker

def check_transaction(amount):
    
    daily_limit = 50000
    
    print("Transaction Amount:", amount)
    
    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")


# Taking user input
transaction_amount = int(input("Enter Transaction Amount: "))

check_transaction(transaction_amount)
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System

def check_exam_eligibility(attendance_list):
    
    total_classes = len(attendance_list)
    attended_classes = 0
    
    # Loop to count attended classes
    for status in attendance_list:
        if status == 1:      # 1 means Present
            attended_classes += 1
    
    attendance_percentage = (attended_classes / total_classes) * 100
    
    print("Attendance Percentage:", attendance_percentage)
    
    if attendance_percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example attendance record (1 = Present, 0 = Absent)
attendance_data = [1, 1, 0, 1, 1]   # 4 out of 5 present = 80%

check_exam_eligibility(attendance_data)
```

```python
# Problem 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    
    # Calculate bill based on slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    
    # Classify usage
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    
    
    return bill, status


# Taking user input
units_consumed = int(input("Enter Units Consumed: "))

total_bill, usage_status = calculate_electricity_bill(units_consumed)

print("Total Bill: ₹", total_bill)
print("Usage Classification:", usage_status)
```
